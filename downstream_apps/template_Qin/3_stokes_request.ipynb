```python
from sunpy.net import Fido
from sunpy.net import attrs as a

# Pick a short time range first (JSOC can stage lots of files)
t = a.Time("2011-01-16T00:00:00", "2011-01-16T01:00:00")

# HMI vector magnetogram, 720s cadence series:
# Requesting hmi.B_720s with magnetic field parameters
q = Fido.search(
    t,
    a.jsoc.Series("hmi.B_720s"),
    a.jsoc.Segment("field"),      # Magnetic field strength
    a.jsoc.Segment("inclination"), # Field inclination angle
    a.jsoc.Segment("azimuth"),     # Field azimuth angle
    a.jsoc.Segment("dop_width"),   # Doppler width
    a.jsoc.Segment("eta_0"),       # Eta_0 parameter
    a.jsoc.Segment("damping"),     # Damping parameter
    a.jsoc.Segment("disambig"), # Doppler shift
    # JSOC requires a notify email for exports/staging:
    a.jsoc.Notify("ql47@njit.edu"),
)

print(q)

files = Fido.fetch(q, path='/home/qinli/surya_workshop/downstream_apps/template_Qin/datasets/hmi/{file}')
print(files)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import glob
import sys
from pathlib import Path

# Add paths for imports
sys.path.append("../../")
sys.path.append("../../Surya")
from magnetogram_utils import synthesize_stokes_single_pixel, plot_stokes_profiles

# Pixel location
pixel_h, pixel_w = 2739, 632

# Path to hmi.B files
hmi_b_folder = '/home/qinli/surya_workshop/downstream_apps/template_Qin/datasets/hmi.B'
date_pattern = '20110116'

# Find all files for the first day
field_files = sorted(glob.glob(f'{hmi_b_folder}/hmi.b_720s.{date_pattern}_*.field.fits'))
inclination_files = sorted(glob.glob(f'{hmi_b_folder}/hmi.b_720s.{date_pattern}_*.inclination.fits'))
azimuth_files = sorted(glob.glob(f'{hmi_b_folder}/hmi.b_720s.{date_pattern}_*.azimuth.fits'))
dop_width_files = sorted(glob.glob(f'{hmi_b_folder}/hmi.b_720s.{date_pattern}_*.dop_width.fits'))
eta_0_files = sorted(glob.glob(f'{hmi_b_folder}/hmi.b_720s.{date_pattern}_*.eta_0.fits'))
damping_files = sorted(glob.glob(f'{hmi_b_folder}/hmi.b_720s.{date_pattern}_*.damping.fits'))

print(f"Found {len(field_files)} sets of files for {date_pattern}")
print(f"Using single pixel at ({pixel_h}, {pixel_w})")

# Extract timestamps and parameter values
times = []
all_stokes = []  # Store synthesized Stokes profiles for each time step
all_wavelengths = None
first_time_params = None  # Store parameters for first time step

for i in range(len(field_files)):
    # Extract timestamp from filename
    filename = field_files[i].split('/')[-1]
    timestamp_str = filename.split('_')[2] + '_' + filename.split('_')[3]  # e.g., '000000_TAI'
    time_str = timestamp_str.split('_')[0]  # e.g., '000000'
    hour = int(time_str[:2])
    minute = int(time_str[2:4])
    second = int(time_str[4:6])
    time_minutes = hour * 60 + minute + second / 60.0
    times.append(time_minutes)
    
    # Load single pixel value
    def load_pixel_value(fits_file):
        with fits.open(fits_file) as hdul:
            data = hdul[1].data if hdul[1].data is not None else hdul[0].data
            return data[pixel_h, pixel_w]
    
    # Load observed parameters at single pixel
    B_mag = load_pixel_value(field_files[i])  # Field strength in Gauss
    theta = np.deg2rad(load_pixel_value(inclination_files[i]))  # Inclination in degrees -> radians
    chi = np.deg2rad(load_pixel_value(azimuth_files[i]))  # Azimuth in degrees -> radians
    dlambdaD = load_pixel_value(dop_width_files[i])  # Doppler width in Angstroms
    eta0 = load_pixel_value(eta_0_files[i])  # Eta_0 parameter
    a = load_pixel_value(damping_files[i])  # Damping parameter
    
    # Convert spherical coordinates to Bx, By, Bz
    Bx = B_mag * np.sin(theta) * np.cos(chi)
    By = B_mag * np.sin(theta) * np.sin(chi)
    Bz = B_mag * np.cos(theta)
    
    # Store parameters for first time step
    if i == 0:
        first_time_params = {
            'Bx': Bx, 'By': By, 'Bz': Bz, 'B_mag': B_mag,
            'theta': theta, 'chi': chi,
            'dlambdaD': dlambdaD, 'eta0': eta0, 'a': a
        }
    
    print(f"\nTime: {time_str} ({time_minutes:.1f} min)")
    print(f"  B = {B_mag:.2f} G, theta = {np.rad2deg(theta):.2f}°, chi = {np.rad2deg(chi):.2f}°")
    print(f"  Bx = {Bx:.2f} G, By = {By:.2f} G, Bz = {Bz:.2f} G")
    print(f"  dlambdaD = {dlambdaD:.4f} Å, eta0 = {eta0:.4f}, a = {a:.4f}")
    
    # Synthesize Stokes profiles using observed ME parameters
    stokes_pred, wavelengths, me_params = synthesize_stokes_single_pixel(
        Bx, By, Bz,
        lambda_rest=6173.15,  # HMI Fe I line
        geff=2.5,  # Effective Landé factor
        eta0=eta0,  # Use observed eta_0
        dlambdaD=dlambdaD,  # Use observed dop_width
        a=a,  # Use observed damping
        lambda0=0.0,  # Assume no velocity shift (can be updated if velocity data available)
        B0=1.0,  # Continuum source function
        B1=0.5,  # Line source function
        device='cuda'  # Use 'cpu' if no GPU available
    )
    
    all_stokes.append(stokes_pred)
    if all_wavelengths is None:
        all_wavelengths = wavelengths
# Convert to numpy arrays
times = np.array(times)
all_stokes = np.array(all_stokes)  # Shape: [n_times, 4, n_wavelengths]

print(f"\nSynthesized {len(times)} time steps")
print(f"Wavelength range: {all_wavelengths.min():.4f} - {all_wavelengths.max():.4f} Å")
print(f"Stokes profiles shape: {all_stokes.shape}")

# Plot synthesized Stokes profiles for the first time step
print(f"\nPlotting synthesized Stokes profiles for first time step (t={times[0]:.1f} min)...")
first_stokes = all_stokes[0]  # [4, n_wavelengths]

plot_stokes_profiles(
    first_stokes,
    all_wavelengths,
    first_time_params['Bx'], 
    first_time_params['By'], 
    first_time_params['Bz'], 
    first_time_params['B_mag'],
    pixel_h=pixel_h,
    pixel_w=pixel_w,
    lambda_rest=6173.15
)
```
