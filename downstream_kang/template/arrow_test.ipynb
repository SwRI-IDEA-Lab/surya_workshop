```python
import pyarrow as pa
import pyarrow.ipc

# Open the file in binary read mode ('rb')
with open("/shared/huggingface_data/surya-bench-flare-forecasting/test/data-00000-of-00001.arrow", "rb") as f:
    # Use open_stream to get a reader object
    reader = pa.ipc.open_stream(f)
    # Read all record batches into a single Table
    table = reader.read_all()
    # You can then convert this table to a pandas DataFrame if needed
    df = table.to_pandas()
    print(df)
```
